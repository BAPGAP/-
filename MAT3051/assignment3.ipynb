# 3.1 #13

```python
import math 

def f(x,n):
    if(n == 1):
        return math.exp(2*x)*math.cos(3*x)
    if(n == 2):
        return math.sin(math.log(x))
    if(n == 3):
        return math.log(x)
    if(n == 4):
        return math.cos(x) + math.sin(x)
    
def atn2(x0,x1,x2,n):
    a0 = f(x0,n)/((x0-x1)*(x0-x2))
    a1 = f(x1,n)/((x1-x0)*(x1-x2))
    a2 = f(x2,n)/((x2-x0)*(x2-x1))

    b0 = x1*x2*a0 + x0*x2*a1 + x0*x1*a2
    b1 = (x1+x2)*a0+(x0+x2)*a1+(x0+x1)*a2
    b1 *= -1
    b2 = a0+a1+a2
    L = [b2, b1, b0]
    return L

def atn3(x0,x1,x2,x3,n):
    a0 = f(x0,n)/((x0-x1)*(x0-x2)*(x0-x3))
    a1 = f(x1,n)/((x1-x0)*(x1-x2)*(x1-x3))
    a2 = f(x2,n)/((x2-x0)*(x2-x1)*(x2-x3))
    a3 = f(x3,n)/((x3-x0)*(x3-x1)*(x3-x2))
    
    b0 = x1*x2*x3*a0 + x0*x2*x3*a1 + x0*x1*x3*a2 + x0*x1*x2*a3
    b0 *= -1
    b1 = (x1*x2+x1*x3+x2*x3)*a0 + (x0*x2+x0*x3+x2*x3)*a1 + (x0*x1+x0*x3+x1*x3)*a2 + (x0*x1+x0*x2+x1*x2)*a3
    b2 = (x1+x2+x3)*a0 + (x0+x2+x3)*a1 + (x0+x1+x3)*a2 + (x0+x1+x2)*a3
    b2 *= -1
    b3 = a0+a1+a2+a3
    L = [b3, b2, b1, b0]
    return L

#13-(a)

(x0, x1, x2) = (0.0, 0.3, 0.6)
print(atn2(x0,x1,x2,1))

#13-(b)

(x0, x1, x2) = (2.0, 2.4, 2.6)
print(atn2(x0,x1,x2,2))

#13-(c)

(x0, x1, x2, x3) = (1,1.1,1.3,1.4)
print(atn3(x0,x1,x2,x3,3))

#13-(d)
(x0, x1, x2, x3) = (0,0.25,0.5,1.0)
print(atn3(x0,x1,x2,x3,4))

```

# 3.4 #9

```python
import numpy as np
x = 10                              #궁금한 x
x_p = [0, 3, 5, 8, 13]              #x 값들
y_p = [0, 225, 383, 623, 993]       #y 값들
y_prime = [75, 77, 80, 74, 72]      #y' 값들
L = 5
z = np.zeros([2*L,2*L+1])
for i in range(L):
    z[2*i][0] = z[2*i+1][0] = x_p[i]
    z[2*i][1] = z[2*i+1][1] = y_p[i]
for i in range(1,2*L):
    if(i%2 == 1):
        m = int(i/2)
        z[i][2] = y_prime[m]
    else:
        z[i][2] = (z[i][1]-z[i-1][1])/(z[i][0]-z[i-1][0])
for j in range(3,2*L+1):
    for i in range(j-1,2*L):
        z[i][j] = (z[i][j-1]-z[i-1][j-1])/(z[i][0]-z[i-j+1][0])
for i in range(2*L):
    for j in range(2*L+1):
        print("%.7f  "%(z[i][j]),end = '')
    print("\n")
    
def f(x):
    answer = 0
    for i in reversed(range(2*L)):
        answer = answer * (x - z[i][0]) + z[i][i+1]
    return answer

print("value of the hermite polynomial at %.1f : %.7f"%(x,f(x)))

def f_p(x,h):
    return (f(x+h)-f(x))/h

def fp(x):
    h = 0.1
    tmp = f_p(x,0.1)
    while(1):
        h /= 10
        pre = tmp
        tmp = f_p(x,h)
        if(abs(tmp-pre) < 1e-4):
            return tmp

print("Speed at 10 : %f"%(fp(10)))

h = 0
mx = 80
mxv = 5
while(h<15):
    tmp = fp(h)
    if(tmp > mx):
        mx = tmp
        mxv = h
    if(tmp>80.6666 and tmp<80.6667):
        print("Speed is %f at %f"%(tmp,h))
    h += 0.0001
print("Maximum Speed is %f at %f"%(mx,mxv))
```

# 3.5 #23

```python
import math,numpy

#초기 값들
n = 4
x = [0, 3, 5, 8, 13]
y = [0, 225, 383, 623, 993]
fpo = 75
fpn = 72

h = numpy.zeros(n+1)
for i in range(n):
    h[i] = x[i+1] - x[i]

a = numpy.zeros(n+1)
a[0] = 3*(y[1]-y[0])/h[0] - 3*fpo
a[n] = 3*fpn - 3*(y[n]-y[n-1])/h[n-1]

for i in range(1,n):
    a[i] = 3/h[i]*(y[i+1]-y[i]) - 3/h[i-1]*(y[i]-y[i-1])
    
l = numpy.zeros(n+1)
z = numpy.zeros(n+1)
m = numpy.zeros(n+1)
b = numpy.zeros(n+1)
c = numpy.zeros(n+1)
d = numpy.zeros(n+1)

(l[0], m[0]) = (2*h[0], 0.5)
z[0] = a[0]/l[0]

for i in range(1,n):
    l[i] = 2*(x[i+1]-x[i-1]) - h[i-1]*m[i-1]
    m[i] = h[i]/l[i]
    z[i] = (a[i]-h[i-1]*z[i-1])/l[i]

l[n] = h[n-1]*(2-m[n-1])
z[n] = (a[n] - h[n-1]*z[n-1])/l[n]
c[n] = z[n]

for j in reversed(range(n)):
    c[j] = z[j] - m[j]*c[j+1]
    b[j] = (y[j+1] - y[j])/h[j] - h[j]*(c[j+1]+2*c[j])/3
    d[j] = (c[j+1]-c[j])/(3*h[j])

for i in range(n):
    print("%dth a,b,c,d : %f %f %f %f"%(i,y[i],b[i],c[i],d[i]))

def f(x,dx):
    m = 0
    x0 = 0
    answer = 0
    if(0 <= x and x <= 3):
        (m, x0) = (0,0)
    elif(3 <= x and x <= 5):
        (m, x0) = (1,3)
    elif(5 <= x and x <= 8):
        (m, x0) = (2,5)
    else:
        (m, x0) = (3,8)
    if(dx):
        LP = [b[m], 2*c[m], 3*d[m]]
        for i in reversed(range(n-1)):
            answer = answer * (x - x0) + LP[i]
        return answer
    
    L = [y[m], b[m], c[m], d[m]]
    for i in reversed(range(n)):
        answer = answer * (x - x0) + L[i]
    return answer

print("value of the cubic spline at 10.0 : %f"%(f(10,False)))
print("Speed at 10 : %f"%(f(10,True)))

h = 0
mx = 80
mxv = 5
ed = False
while(h<13):
    tmp = f(h,True)
    if(tmp > mx):
        mx = tmp
        mxv = h
    if(tmp>80.66666 and tmp<80.66667 and not ed):
        print("Speed is %f at %f"%(tmp,h))
        ed = True
    h += 0.0001
print("Maximum Speed is %f at %f"%(mx,mxv))
```

# Algorithm 3.5 (Clamped Cubic Spline)

```python
import math,numpy

#초기 값들
n = 3
e = math.exp(1)
x = [0, 1, 2, 3]
y = [1, e, e*e, e*e*e]
fpo = 1
fpn = e*e*e

h = numpy.zeros(n+1)
for i in range(n):
    h[i] = x[i+1] - x[i]

a = numpy.zeros(n+1)
a[0] = 3*(y[1]-y[0])/h[0] - 3*fpo
a[n] = 3*fpn - 3*(y[n]-y[n-1])/h[n-1]

for i in range(1,n):
    a[i] = 3/h[i]*(y[i+1]-y[i]) - 3/h[i-1]*(y[i]-y[i-1])
    
l = numpy.zeros(n+1)
z = numpy.zeros(n+1)
m = numpy.zeros(n+1)
b = numpy.zeros(n+1)
c = numpy.zeros(n+1)
d = numpy.zeros(n+1)

(l[0], m[0]) = (2*h[0], 0.5)
z[0] = a[0]/l[0]

for i in range(1,n):
    l[i] = 2*(x[i+1]-x[i-1]) - h[i-1]*m[i-1]
    m[i] = h[i]/l[i]
    z[i] = (a[i]-h[i-1]*z[i-1])/l[i]

l[n] = h[n-1]*(2-m[n-1])
z[n] = (a[n] - h[n-1]*z[n-1])/l[n]
c[n] = z[n]

for j in reversed(range(n)):
    c[j] = z[j] - m[j]*c[j+1]
    b[j] = (y[j+1] - y[j])/h[j] - h[j]*(c[j+1]+2*c[j])/3
    d[j] = (c[j+1]-c[j])/(3*h[j])

for i in range(n):
    print("%dth a,b,c,d : %f %f %f %f"%(i,y[i],b[i],c[i],d[i]))
```

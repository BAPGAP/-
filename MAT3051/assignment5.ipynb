```python
#5.2 - 9
import numpy
t = numpy.arange(1,2.1,0.1)
y = numpy.zeros(11)
e = numpy.zeros(11)
h = 0.1

def r(t):
    return t*t*(numpy.exp(t)-numpy.exp(1))

def f(t,y):
    return 2*y/t+t*t*numpy.exp(t)

for i in range(1,11):
    y[i] = y[i-1] + h * f(t[i-1],y[i-1])
    e[i] = r(t[i])

print(y)
print(e)
print((y[1]-y[0])*(0.04)/0.1+y[0],r(1.04))
print((y[6]-y[5])*(0.05)/0.1+y[5],r(1.55))
print((y[10]-y[9])*(0.07)/0.1+y[9],r(1.97))
```

```python
#5.3 - 10
import numpy
t = numpy.arange(1,2.05,0.05)
y = numpy.zeros(21)
y_2 = numpy.zeros(21)
y_4 = numpy.zeros(21)
h = 0.05
y_4[0] = y_2[0] = y[0] = -1

def f(t,y,n):
    if(n==0):
        return 1/(t*t)-y/t-y*y
    elif(n==1):
        return -3/(t*t*t)+3*y*y/t+2*y*y*y
    elif(n==2):
        return 9/(t*t*t*t)+6*y/(t*t*t)-3*y*y/(t*t)-12*y*y*y/t-6*y*y*y*y
    elif(n==3):
        return -30/(t*t*t*t*t)*(1+t*y+t*t*y*y-t*t*t*y*y*y)+60*y*y*y*y/t+24*y*y*y*y*y

def f_exact(t):
    return -1/t

def T(t,y,n):
    if(n==2):
        return f(t,y,0)+h/2*f(t,y,1)
    elif(n==4):
        return f(t,y,0)+h/2*f(t,y,1)+h*h/6*f(t,y,2)+h*h*h/24*f(t,y,3)

for i in range(1,21):
    y[i] = f_exact(t[i])
    y_2[i] = y_2[i-1] + h * T(t[i-1],y_2[i-1],2)
    y_4[i] = y_4[i-1] + h * T(t[i-1],y_4[i-1],4)
    
for i in range(0,21):
    print(y[i], y_2[i], y_4[i])
    
def f_l(gap):
    tmp = (int)(gap*100)
    tmp = tmp - tmp%5
    tmp = (int)((tmp - 100)/5) #tmp ~ tmp+1
    return (y_2[tmp+1] - y_2[tmp])*(gap-t[tmp])/0.05 + y_2[tmp]
    
def f_c(gap):
    tmp = (int)(gap*100)
    tmp = tmp - tmp%5
    tmp = (int)((tmp - 100)/5) #tmp ~ tmp+1
    
    a1 = f(t[tmp],y_4[tmp],0)
    a2 = (y_4[tmp+1]-y_4[tmp])/0.05
    a3 = f(t[tmp+1],y_4[tmp+1],0)
    b1 = (a2-a1)/0.05
    b2 = (a3-a2)/0.05
    c1 = (b2-b1)/0.05
    
    return y_4[tmp]+(gap-t[tmp])*a1+(gap-t[tmp])*(gap-t[tmp])*b1+(gap-t[tmp])*(gap-t[tmp])*(gap-t[tmp+1])*c1

print("\n")
print(f_exact(1.052),f_l(1.052),f_c(1.052))
print(f_exact(1.555),f_l(1.555),f_c(1.555))
print(f_exact(1.978),f_l(1.978),f_c(1.978))
    
```
